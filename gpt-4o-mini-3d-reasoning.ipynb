```python
import random
import ast
import pandas as pd
import numpy as np


def safe_eval(val):
    if isinstance(val, list):  # already list
        return val
    if pd.isna(val):  # NaN
        return []
    if isinstance(val, str):  # string that looks like a list
        return ast.literal_eval(val)
    return []  # fallback

df['correct_combinations'] = df['correct_combinations'].apply(safe_eval)
df['incorrect_combinations'] = df['incorrect_combinations'].apply(safe_eval)


qa_pairs = []
correct_label_cycle = ['A', 'B', 'C', 'D']
label_index = 0  # To cycle through correct answer positions


for idx, row in df.iterrows():
    if len(row['correct_combinations']) == 0 or len(row['incorrect_combinations']) == 0:
        continue  # skip if missing answers
    
    question = row['question']
    # if len(row['correct_combinations']) > 1:
    #     correct_ans = row['correct_combinations'][1]
    # else:
    correct_ans = row['correct_combinations'][0]
    /
    incorrect_list = row['incorrect_combinations']
    if len(incorrect_list) >= 3:
        incorrect_ans = random.sample(incorrect_list, 3)
    else:
        incorrect_ans = random.choices(incorrect_list, k=3)
    
    # We'll fix the correct answer position according to label_index
    labels = ['A', 'B', 'C', 'D']
    correct_label = correct_label_cycle[label_index % 4]
    label_index += 1
    
    # Place correct answer in the position indicated by correct_label
    options = incorrect_ans.copy()
    options.insert(labels.index(correct_label), correct_ans)
    
    qa_pairs.append({
        'question': question,
        'A': options[0],
        'B': options[1],
        'C': options[2],
        'D': options[3],
        'correct_label': correct_label,
        'correct_answer_text': correct_ans
    })

qa_df = pd.DataFrame(qa_pairs)
print(f"Generated {len(qa_df)} QA pairs with balanced correct answer positions.")

qa_df.to_csv('csqa_pairs.csv',index=False)
```

```python
import os
import json
import time
import pandas as pd
import random
from openai import OpenAI, RateLimitError
from tenacity import retry, wait_random_exponential, stop_after_attempt
from datasets import load_dataset
from kaggle_secrets import UserSecretsClient
user_secrets = UserSecretsClient()

# --- 1. Setup and API Call Functions ---

openai_key = user_secrets.get_secret("openai")
if not openai_key:
    raise ValueError("OPENAI_API_KEY environment variable not set.")
client = OpenAI(api_key=openai_key)


def call_gpt4o_mini(messages):
    """Sends a prompt to the GPT-4o-mini API with robustness for errors."""
    try:
        response = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=messages,
            response_format={"type": "json_object"},
            temperature=0,
            n=1
        )
        content = response.choices[0].message.content
        return json.loads(content)
    except RateLimitError as e:
        print(f"Rate limit error: {e}. Retrying...")
        raise
    except json.JSONDecodeError as e:
        print(f"Received invalid JSON from API: {e}. Retrying...")
        raise e
    except Exception as e:
        print(f"An unexpected error occurred during the API call: {e}")
        raise e


def one_shot_three_depth_reasoning(row):
    """
    Creates a prompt for a three-depth reasoning task with nested premises.
    
    The prompt instructs the model to analyze each option by:
    1.  Starting with a top-level Hypothesis (H).
    2.  Breaking it down into two Sub-Hypotheses (H1 and H2).
    3.  For each Sub-Hypothesis, analyzing two specific actions (e.g., A1.1 and A1.2).
    4.  Providing a two-premise breakdown for each action.
    5.  The conclusion for each Sub-Hypothesis is explicitly stated as being entailed by the premises.
    6.  Finally, evaluating the AND logic and providing a definitive 'Therefore' statement.
    """
    user_content = f"""
    You are a logical reasoning AI designed to evaluate complex, multi-factor, multiple-choice questions. To find the correct answer, you must analyze each option and determine if both parts of the "AND" combination positively contribute to the overall goal.

    Your task is to analyze each option and provide a structured, multi-step breakdown. For each option, you will:

    1.  Start with a top-level Hypothesis (H) that reflects the overall goal of the question.
    
    2.  Break down the hypothesis into two Sub-Hypotheses (H1 and H2) that correspond to the two factors mentioned in the question.
    
    3.  For each Sub-Hypothesis, analyze the specific Actions (A1.1, A1.2, etc.) taken in the option.
    
    4.  Provide a two-premise breakdown for each action, detailing the logical steps.
    
    5.  The conclusion for each Sub-Hypothesis is explicitly stated as being entailed by the premises.
    
    6.  Provide an AND Logic Evaluation that summarizes how the two top-level actions work together.
    
    7.  Finally, provide a concluding Therefore statement that definitively judges the option's effectiveness.

    8.  At the end, provide the right choice (just the letter, e.g., 'A', and not the full sentence).
    
    Below is an example for your reference. Note the three-depth logical breakdown for each action and the new conclusion field that represents the entailment.
    
    Question: What are ways someone can get warm?
    A. Wearing thick clothing and turning off the lights
    B. Drinking hot tea and sitting near an open window
    C. Turning on a heater and closing windows
    D. Eating spicy food and taking a cold shower
    
    Output: {{
        "output_breakdown": [
            {{
                "option": "A",
                "hypothesis": "The person stays warm in extremely cold weather",
                "sub_hypotheses": [
                    {{
                        "id": "H1",
                        "factor": "The body retains heat",
                        "actions": [
                            {{
                                "id": "A1.1",
                                "description": "The person wears thick clothing",
                                "premises": [
                                    {{
                                        "id": "A1.1.1",
                                        "text": "Thick clothes act as insulation by trapping air."
                                    }},
                                    {{
                                        "id": "A1.1.2",
                                        "text": "Insulation slows heat loss from the body."
                                    }}
                                ]
                            }},
                            {{
                                "id": "A1.2",
                                "description": "Heat stays close to the body",
                                "premises": [
                                    {{
                                        "id": "A1.2.1",
                                        "text": "Insulated air layers reduce heat transfer from the body to the environment."
                                    }},
                                    {{
                                        "id": "A1.2.2",
                                        "text": "Reduced heat transfer helps maintain body temperature."
                                    }}
                                ]
                            }}
                        ],
                        "conclusion": "Therefore, the body effectively retains heat due to the insulating properties of thick clothing."
                    }},
                    {{
                        "id": "H2",
                        "factor": "External cold is blocked",
                        "actions": [
                            {{
                                "id": "A2.1",
                                "description": "The person turns off the lights",
                                "premises": [
                                    {{
                                        "id": "A2.1.1",
                                        "text": "Lights produce a negligible amount of heat."
                                    }},
                                    {{
                                        "id": "A2.1.2",
                                        "text": "Turning them off has no significant effect on room temperature."
                                    }}
                                ]
                            }},
                            {{
                                "id": "A2.2",
                                "description": "No cold air is blocked",
                                "premises": [
                                    {{
                                        "id": "A2.2.1",
                                        "text": "Cold drafts and heat loss still occur without a physical barrier."
                                    }},
                                    {{
                                        "id": "A2.2.2",
                                        "text": "Heat is lost even when lights are off."
                                    }}
                                ]
                            }}
                        ],
                        "conclusion": "Therefore, external cold is not blocked because turning off the lights is irrelevant to thermal insulation or blocking drafts."
                    }}
                ],
                "and_logic_evaluation": "Wearing thick clothing helps retain warmth, BUT turning off the lights is irrelevant to blocking external cold.",
                "therefore": "This combination is not the best choice because one of the actions is ineffective."
            }},
            {{
                "option": "B",
                "hypothesis": "The person stays warm in extremely cold weather",
                "sub_hypotheses": [
                    {{
                        "id": "H1",
                        "factor": "The person increases internal warmth",
                        "actions": [
                            {{
                                "id": "B1.1",
                                "description": "The person drinks hot tea",
                                "premises": [
                                    {{
                                        "id": "B1.1.1",
                                        "text": "Hot tea briefly warms internal body temperature."
                                    }},
                                    {{
                                        "id": "B1.1.2",
                                        "text": "Warm liquids promote circulation, which can help distribute warmth."
                                    }}
                                ]
                            }},
                            {{
                                "id": "B1.2",
                                "description": "Internal warmth can reduce cold impact",
                                "premises": [
                                    {{
                                        "id": "B1.2.1",
                                        "text": "A higher internal temperature delays the onset of shivering."
                                    }},
                                    {{
                                        "id": "B1.2.2",
                                        "text": "Better circulation helps distribute warmth throughout the body."
                                    }}
                                ]
                            }}
                        ],
                        "conclusion": "Therefore, drinking hot tea provides a temporary increase in internal warmth, which helps reduce the impact of the cold."
                    }},
                    {{
                        "id": "H2",
                        "factor": "Cold exposure is avoided",
                        "actions": [
                            {{
                                "id": "B2.1",
                                "description": "The person sits near an open window",
                                "premises": [
                                    {{
                                        "id": "B2.1.1",
                                        "text": "Cold air enters through the open window."
                                    }},
                                    {{
                                        "id": "B2.1.2",
                                        "text": "Cold drafts lower the ambient room temperature."
                                    }}
                                ]
                            }},
                            {{
                                "id": "B2.2",
                                "description": "Cold air exposure causes heat loss",
                                "premises": [
                                    {{
                                        "id": "B2.2.1",
                                        "text": "Drafts cool the body directly, causing discomfort."
                                    }},
                                    {{
                                        "id": "B2.2.2",
                                        "text": "Body heat escapes faster in cold, moving air."
                                    }}
                                ]
                            }}
                        ],
                        "conclusion": "Therefore, cold exposure is not avoided, and the person's body will lose heat faster."
                    }}
                ],
                "and_logic_evaluation": "Drinking hot tea provides temporary warmth, BUT sitting near an open window causes significant heat loss, directly contradicting the goal.",
                "therefore": "This combination is counterproductive and not the best choice."
            }},
            {{
                "option": "C",
                "hypothesis": "The person stays warm in extremely cold weather",
                "sub_hypotheses": [
                    {{
                        "id": "H1",
                        "factor": "The room is kept warm",
                        "actions": [
                            {{
                                "id": "C1.1",
                                "description": "The person turns on a heater",
                                "premises": [
                                    {{
                                        "id": "C1.1.1",
                                        "text": "Heaters convert electrical energy into thermal energy."
                                    }},
                                    {{
                                        "id": "C1.1.2",
                                        "text": "The warmer air increases the rooms temperature."
                                    }}
                                ]
                            }},
                            {{
                                "id": "C1.2",
                                "description": "The warm room helps maintain body heat",
                                "premises": [
                                    {{
                                        "id": "C1.2.1",
                                        "text": "A warm environment reduces body heat loss to the surrounding air."
                                    }},
                                    {{
                                        "id": "C1.2.2",
                                        "text": "Less heat loss helps the person stay warm and comfortable."
                                    }}
                                ]
                            }}
                        ],
                        "conclusion": "Therefore, the room is effectively kept warm, which helps the body retain heat."
                    }},
                    {{
                        "id": "H2",
                        "factor": "Cold air from outside is blocked",
                        "actions": [
                            {{
                                "id": "C2.1",
                                "description": "The person closes the windows",
                                "premises": [
                                    {{
                                        "id": "C2.1.1",
                                        "text": "Open windows allow cold drafts to enter the room."
                                    }},
                                    {{
                                        "id": "C2.1.2",
                                        "text": "Cold drafts lower the room temperature."
                                    }}
                                ]
                            }},
                            {{
                                "id": "C2.2",
                                "description": "No drafts means indoor heat is retained",
                                "premises": [
                                    {{
                                        "id": "C2.2.1",
                                        "text": "Closing windows stops cold air inflow."
                                    }},
                                    {{
                                        "id": "C2.2.2",
                                        "text": "A stable room temperature helps maintain warmth and reduce drafts."
                                    }}
                                ]
                            }}
                        ],
                        "conclusion": "Therefore, cold air from outside is effectively blocked, preventing heat loss."
                    }}
                ],
                "and_logic_evaluation": "A heater provides heat generation, AND closed windows prevent that heat from escaping. Both actions work together to maximize warmth.",
                "therefore": "This combination is the correct choice because both actions are effective and synergistic."
            }},
            {{
                "option": "D",
                "hypothesis": "The person stays warm in extremely cold weather",
                "sub_hypotheses": [
                    {{
                        "id": "H1",
                        "factor": "The person raises body temperature",
                        "actions": [
                            {{
                                "id": "D1.1",
                                "description": "The person eats spicy food",
                                "premises": [
                                    {{
                                        "id": "D1.1.1",
                                        "text": "Spicy food causes a sensation of warmth by stimulating nerve receptors."
                                    }},
                                    {{
                                        "id": "D1.1.2",
                                        "text": "The sensation is a neural response, not an increase in core body temperature."
                                    }}
                                ]
                            }},
                            {{
                                "id": "D1.2",
                                "description": "No lasting heat is produced",
                                "premises": [
                                    {{
                                        "id": "D1.2.1",
                                        "text": "Spicy food can cause sweating, which is the body's natural cooling mechanism."
                                    }},
                                    {{
                                        "id": "D1.2.2",
                                        "text": "The net result of sweating is body heat loss, not gain."
                                    }}
                                ]
                            }}
                        ],
                        "conclusion": "Therefore, the person does not effectively raise their body temperature and may even lose heat from sweating."
                    }},
                    {{
                        "id": "H2",
                        "factor": "Cold exposure is reduced",
                        "actions": [
                            {{
                                "id": "D2.1",
                                "description": "The person takes a cold shower",
                                "premises": [
                                    {{
                                        "id": "D2.1.1",
                                        "text": "Cold water draws heat away from skin and internal organs."
                                    }},
                                    {{
                                        "id": "D2.1.2",
                                        "text": "Skin temperature drops rapidly, triggering the body to preserve heat."
                                    }}
                                ]
                            }},
                            {{
                                "id": "D2.2",
                                "description": "Shower increases cold exposure",
                                "premises": [
                                    {{
                                        "id": "D2.2.1",
                                        "text": "Wet skin loses heat much faster than dry skin."
                                    }},
                                    {{
                                        "id": "D2.2.2",
                                        "text": "The cold water in a shower increases heat loss rather than reducing cold exposure."
                                    }}
                                ]
                            }}
                        ],
                        "conclusion": "Therefore, cold exposure is significantly increased, leading to rapid heat loss."
                    }}
                ],
                "and_logic_evaluation": "Eating spicy food provides a minimal, psychological sense of warmth, BUT a cold shower causes significant heat loss.",
                "therefore": "This combination is counterproductive and not the best choice."
            }}
        ],
        "final_answer": "C"
    }}
    
    Now your turn:
    Question:
    {str(row['question'])}
    
    Options:
    A. {str(row['A'])}
    B. {str(row['B'])}
    C. {str(row['C'])}
    D. {str(row['D'])}
    
    Generate the response as a JSON object, following the exact example format.
    Output: {{
      "output_breakdown": [
        {{
          "option": "A",
          "hypothesis": "...",
          "sub_hypotheses": [
            {{
              "id": "H1",
              "factor": "...",
              "actions": [
                {{
                  "id": "A1.1",
                  "description": "...",
                  "premises": [
                    {{
                      "id": "A1.1.1",
                      "text": "..."
                    }},
                    {{
                      "id": "A1.1.2",
                      "text": "..."
                    }}
                  ]
                }},
                {{
                  "id": "A1.2",
                  "description": "...",
                  "premises": [
                    {{
                      "id": "A1.2.1",
                      "text": "..."
                    }},
                    {{
                      "id": "A1.2.2",
                      "text": "..."
                    }}
                  ]
                }}
              ],
              "conclusion": "..."
            }},
            {{
              "id": "H2",
              "factor": "...",
              "actions": [
                {{
                  "id": "A2.1",
                  "description": "...",
                  "premises": [
                    {{
                      "id": "A2.1.1",
                      "text": "..."
                    }},
                    {{
                      "id": "A2.1.2",
                      "text": "..."
                    }}
                  ]
                }},
                {{
                  "id": "A2.2",
                  "description": "...",
                  "premises": [
                    {{
                      "id": "A2.2.1",
                      "text": "..."
                    }},
                    {{
                      "id": "A2.2.2",
                      "text": "..."
                    }}
                  ]
                }}
              ],
              "conclusion": "..."
            }}
          ],
          "and_logic_evaluation": "...",
          "therefore": "..."
        }},
        {{
          "option": "B",
          "hypothesis": "...",
          "sub_hypotheses": [
            {{
              "id": "H1",
              "factor": "...",
              "actions": [
                {{
                  "id": "B1.1",
                  "description": "...",
                  "premises": [
                    {{
                      "id": "B1.1.1",
                      "text": "..."
                    }},
                    {{
                      "id": "B1.1.2",
                      "text": "..."
                    }}
                  ]
                }},
                {{
                  "id": "B1.2",
                  "description": "...",
                  "premises": [
                    {{
                      "id": "B1.2.1",
                      "text": "..."
                    }},
                    {{
                      "id": "B1.2.2",
                      "text": "..."
                    }}
                  ]
                }}
              ],
              "conclusion": "..."
            }}
          ],
          "and_logic_evaluation": "...",
          "therefore": "..."
        }},
        {{
          "option": "C",
          "hypothesis": "...",
          "sub_hypotheses": [
            {{
              "id": "H1",
              "factor": "...",
              "actions": [
                {{
                  "id": "C1.1",
                  "description": "...",
                  "premises": [
                    {{
                      "id": "C1.1.1",
                      "text": "..."
                    }},
                    {{
                      "id": "C1.1.2",
                      "text": "..."
                    }}
                  ]
                }},
                {{
                  "id": "C1.2",
                  "description": "...",
                  "premises": [
                    {{
                      "id": "C1.2.1",
                      "text": "..."
                    }},
                    {{
                      "id": "C1.2.2",
                      "text": "..."
                    }}
                  ]
                }}
              ],
              "conclusion": "..."
            }}
          ],
          "and_logic_evaluation": "...",
          "therefore": "..."
        }},
        {{
          "option": "D",
          "hypothesis": "...",
          "sub_hypotheses": [
            {{
              "id": "H1",
              "factor": "...",
              "actions": [
                {{
                  "id": "D1.1",
                  "description": "...",
                  "premises": [
                    {{
                      "id": "D1.1.1",
                      "text": "..."
                    }},
                    {{
                      "id": "D1.1.2",
                      "text": "..."
                    }}
                  ]
                }},
                {{
                  "id": "D1.2",
                  "description": "...",
                  "premises": [
                    {{
                      "id": "D1.2.1",
                      "text": "..."
                    }},
                    {{
                      "id": "D1.2.2",
                      "text": "..."
                    }}
                  ]
                }}
              ],
              "conclusion": "..."
            }}
          ],
          "and_logic_evaluation": "...",
          "therefore": "..."
        }}
      ],
      "final_answer": ""
    }}
    """
    
    messages = [
    {"role": "system", "content": "You are a logical reasoning AI designed to evaluate complex, multi-factor, multiple-choice questions. To find the correct answer, you must analyze each option and determine if both parts of the 'AND' combination positively contribute to the overall goal of the question. Your task is to analyze each option and provide a structured, multi-step breakdown."},
    {"role": "user", "content": user_content}
    ]
    return messages


def process_single_row(qa_df, output_file_csv):
    """
    Orchestrates the single-stage pipeline and saves the final output to a CSV file.
    """
    print(f"Starting single-stage pipeline for {len(qa_df)} rows.")
    final_output_list = []
    
    required_cols = ['question', 'A', 'B', 'C', 'D', 'correct_label']
    if not all(col in qa_df.columns for col in required_cols):
        raise ValueError(f"DataFrame must contain the following columns: {required_cols}")

    valid_choices = {'A', 'B', 'C', 'D'}

    for idx, row in qa_df.iterrows():
        # This is where the model generates the question
        messages = one_shot_three_depth_reasoning(row)
        try:
            # Call the API and get the full JSON output
            api_result_dict = call_gpt4o_mini(messages)
            
            
            # Extract the reasoning and final answer
            generated_reasoning = json.dumps(api_result_dict.get('output_breakdown', []))
            predicted_label = api_result_dict.get('final_answer')
            
            # Create the final result dictionary for this row
            result = {
                'question': row['question'],
                'options': f"A. {row['A']} B. {row['B']} C. {row['C']} D. {row['D']}",
                'reasoning_full': generated_reasoning,
                'predicted_label': predicted_label.strip().upper() if predicted_label and predicted_label.strip().upper() in valid_choices else None,
                'correct_label': row['correct_label']
            }

            final_output_list.append(result)
            print(f"Processed row {idx} successfully. Predicted: {result.get('predicted_label')}, Correct: {result.get('correct_label')}")
            
        except Exception as e:
            print(f"Failed to process row {idx}. Error: {e}. Skipping.")
            final_output_list.append({
                'question': row['question'],
                'options': f"A. {row['A']} B. {row['B']} C. {row['C']} D. {row['D']}",
                'reasoning_full': "",
                'predicted_label': None,
                'correct_label': row['correct_label']
            })

    final_df = pd.DataFrame(final_output_list)
    final_df.to_csv(output_file_csv, index=False, encoding='utf-8')
    print(f"\nProcessing complete. {len(final_output_list)} final results saved to '{output_file_csv}'.")

# --- Example of running the pipeline ---
if __name__ == '__main__':
    process_single_row(qa_df.iloc[:200], 'csqa_oneshot_3d_reasoning.csv')
```
