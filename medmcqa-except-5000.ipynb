```python
import json
paths=[]
files=[]
for dirname, _, filenames in os.walk('/kaggle/input/medmcqa-database/'):
    for filename in filenames:
        files+=[filename]
        paths+=[(os.path.join(dirname, filename))]
DBs=paths.copy()
for i in range(len(paths)):
    file=paths[i].split('/')[-1]
    print(i,file)
    db=[]
    with open(paths[i], 'r') as file:
        for line in file:
            db.append(json.loads(line))
    print(len(db))
    DBs[i]=pd.DataFrame(db)
print(len(DBs))

ddf0=DBs[0]
df0=DBs[1]
tdf0=DBs[2]

df=pd.concat([df0,ddf0, tdf0])

print(df.columns.tolist())

df=df[['id','question', 'opa', 'opb', 'opc', 'opd','cop', 'choice_type' ]]
df.columns=['id','question', 'A','B','C','D','answer','choice_type']
df['answer']=df['answer'].map({1:'A',2:'B',3:'C',4:'D'})
```

```python
negation_keywords = ["except"]
all_questions = df.question.to_list()
def has_negation(q):
    q_lower = q.lower()
    return any(word in q_lower for word in negation_keywords)

filtered_questions = [q for q in all_questions if has_negation(q)]
```

```python
import pandas as pd
import json
import os
from openai import OpenAI

def modify_dataset_with_openai(client, qa_dataframe):
    """
    Performs zero-shot question answering on a DataFrame of multiple-choice questions
    using the OpenAI GPT-4o-mini API.

    Args:
        client: An initialized OpenAI client.
        qa_dataframe: A pandas DataFrame with columns 'question', 'A', 'B', 'C', 'D',
                      and 'correct_label'.

    Returns:
        A list of dictionaries, where each dictionary contains the original question,
        the model's transformed output, and the correct label.
    """
    results = []
    # System message containing the detailed instructions for the model
    system_prompt = """
    You are a medical question transformation assistant who is creating a questionnaire for an exam.
    """
    i = 1
    for idx, row in qa_dataframe.iterrows():
        i+=1
        if i % 100 == 0:
            print(i)
        try:
            # User message containing the specific question and options
            user_prompt = f"""
You are an expert medical question writer tasked with converting single-choice questions into "Select all that apply" questions. Your goal is to logically **invert** the question's premise so that multiple options become correct.

Here are the rules you must follow:

1.  **Invert the question's logic.**
    * If the original question asks for the **single correct** answer, rephrase it to ask for **all correct** answers.
    * If the original question asks for the **single incorrect** answer (e.g., uses "except," "not," or "false"), rephrase it to ask for **all correct** answers.
    * The new `correct_answers` list must contain all options that are logically consistent with the transformed question.

2.  **Use the exact JSON output format.** The output must be a single JSON object with the following three keys:
    * `"transformed_question"`: The rewritten question text.
    * `"options"`: An object containing the original options, keyed by their letters (A, B, C, D).
    * `"correct_answers"`: A list of the letters of all correct options for the new question.

3.  **Maintain the original options.** Do not add or remove options.

Example:

Original Question: All of the following are surgical options for morbid obesity except -
A: Adjustable gastric banding
B: Biliopancreatic diversion
C: Duodenal Switch
D: Roux en Y Duodenal Bypass
Correct Answer: D

Output:
{{
  "transformed_question": "Which of the following are surgical options for morbid obesity? Select all that apply.",
  "options": {{
      "A": "Adjustable gastric banding",
      "B": "Biliopancreatic diversion",
      "C": "Duodenal Switch",
      "D": "Roux en Y Duodenal Bypass"
  }},
  "correct_answers": ["A", "B", "C"]
}}

Now transform this question:

Original Question: {row['question']}
Options: A: {row['A']} B: {row['B']} C: {row['C']} D: {row['D']}
Correct Answer: {row['answer']}
"""

            
            
            # Make the API call to GPT-4o-mini
            completion = client.chat.completions.create(
                model="gpt-4o-mini",
                messages=[
                    {"role": "system", "content": system_prompt},
                    {"role": "user", "content": user_prompt}
                ],
                response_format={"type": "json_object"}
            )
            
            # Parse the JSON from the API response
            generated_text = completion.choices[0].message.content
            parsed_output = json.loads(generated_text)
            
            # Append results
            results.append({
                "original_question": row['question'],
                "original_options": {
                    "A": row['A'],
                    "B": row['B'],
                    "C": row['C'],
                    "D": row['D']
                },
                "original_answer": row['answer'],
                "transformed": parsed_output
            })

        except Exception as e:
            results.append({
                "original_question": row['question'],
                "error": str(e)
            })

    return results

# Example usage with a dummy DataFrame
# You need to set your OpenAI API key in your environment variables
# export OPENAI_API_KEY="YOUR_API_KEY"

# Ensure you have the 'openai' library installed: pip install openai
client = OpenAI(api_key=user_secrets.get_secret("openai"))
filtered_set = set(filtered_questions) 
filtered_df = df[df['question'].isin(filtered_set)]
results = modify_dataset_with_openai(client, filtered_df.iloc[:5000])
```

```python
my_results = pd.DataFrame(results)
my_results.to_csv('medmcqa_dataset.csv', index=False)
```
